Question 1: What is Simple Linear Regression?
Ans:
imple linear regression is a statistical method that models the relationship between one dependent variable (y) and one independent variable (x) using a straight line. It finds the "best-fit" line—defined by the equation y=beta0 +beta1x-e to predict outcomes or understand how one variable changes with another.

Core Concepts
Dependent Variable (y): The outcome or output variable you are trying to predict or explain (e.g., salary).
Independent Variable (x): The predictor or input variable used to explain changes in
 (e.g., years of experience).
The Best Fit Line (y=beta0+beta1x):

 beta0(Intercept): The point where the line crosses the y-axis.

 beta1(Slope): The steepness of the line, indicating how much
 changes for a one-unit increase in x.
Residuals (Errors): The vertical distance between the actual data points and the regression line.

How It Works
Simple linear regression uses a method called Ordinary Least Squares (OLS). This technique finds the specific line that minimizes the sum of all squared residual distances between the data points and the regression line.

Key Assumptions
For the model to be valid, it should meet several criteria:

Linearity: The relationship between
 X and Y is linear.
Independence: Observations are independent of each other.
Homoscedasticity: The residuals have a constant variance (constant spread across the line).
Normality: The errors (residuals) are normally distributed.


Question 2: What are the key assumptions of Simple Linear Regression?
aNS:
The key assumptions of Simple Linear Regression (often remembered by the acronym LINE) are Linearity between variables, Independence of observations, Normality of errors, and Equal Variance (homoscedasticity) of residuals. These assumptions ensure the model is reliable and the estimations are unbiased.
Linearity (L): A linear relationship exists between the independent variable (X) and the dependent variable (Y), meaning a straight line can describe the relationship.
Independence of Residuals/Errors (I): Observations are independent of one another. The residuals are not correlated with each other (crucial for time-series data).
Normality (N): The residuals (errors) follow a normal distribution, which is especially critical for valid hypothesis testing and building confidence intervals.
Equality of Variance (E) / Homoscedasticity: The residuals have consistent variance across all levels of the independent variable, meaning the error terms do not widen (fan out) as X
 changes.
Additional Assumptions:
No Endogeneity: The independent variable is not correlated with the error term (exogeneity).
No Perfect Multicollinearity: While usually for multiple regression, this implies in simple regression that the predictor (X) has variance and is not constant.
Checking Assumptions:
These assumptions are typically verified by examining residual plots. For example, a random scatter of residuals confirms homoscedasticity, while a pattern indicates a violation.


Question 3: What is heteroscedasticity, and why is it important to address in regression
models?
Ans:Heteroscedasticity occurs in regression analysis when the variance of the residuals (errors) is not constant across all levels of the independent variables, creating a "fan" or "cone" shape in residual plots. It is crucial to address because it violates Ordinary Least Squares (OLS) assumptions, causing inefficient coefficient estimates, biased standard errors, and invalid hypothesis tests (p-values).
Why Addressing Heteroscedasticity is Important
Invalid Hypothesis Tests: While OLS coefficients remain unbiased, the standard errors are biased, which invalidates confidence intervals and
-values.
Inefficient Estimates: The regression estimates are no longer efficient (i.e., they are not the best linear unbiased estimators), meaning they do not provide the most precise prediction.
Incorrect Inferences: Overconfidence in a model can lead to false conclusions about the significance of predictor variables.

Common Causes and Detection
Causes: Heteroscedasticity often arises in cross-sectional data where variables vary significantly in scale (e.g., income vs. expenditure). It can also result from a "learning effect" (variance decreases as behavior becomes more consistent) or from using averaged data.
Detection: It is identified by analyzing residual plots for a non-constant, widening, or "fan" pattern. Formal tests like the White test or Breusch-Pagan test can also be used.


How to Address It
Data Transformation: Applying logarithmic transformations (
) to the dependent variable can shrink the variance.
Weighted Least Squares (WLS): This method assigns smaller weights to observations with higher variance, improving efficiency.
Robust Standard Errors: Using White’s heteroscedasticity-consistent standard errors allows for valid inference without transforming the data


Question 4: What is Multiple Linear Regression?
Ans:

Multiple linear regression (MLR) is a statistical technique that models the relationship between one continuous dependent variable and two or more independent (predictor) variables by fitting a linear equation to observed data. It predicts outcomes based on multiple factors, assuming a linear relationship, minimal multicollinearity, and constant variance of residuals.

Independent Variables: These can be continuous or categorical.
Key Assumptions:
Linearity: The relationship between predictors and the dependent variable is linear.
No Multicollinearity: Independent variables should not be too highly correlated.
Homoscedasticity: The variance of residual terms is constant.
Independence: Observations are independent of each other.
Example Use Case:
Predicting a house's price (dependent variable) based on factors like square footage, number of bedrooms, and location (independent variables).

Benefits:
Enhanced Accuracy: Incorporating multiple factors usually provides better predictive power than simple linear regression.
Relative Influence: It helps determine the relative contribution of each independent variable to the total variance explained.

Common Pitfalls:
Multicollinearity: If independent variables are highly correlated, it becomes difficult to determine the individual effect of each variable.
Overfitting: Including too many variables can lead to a model that performs well on training data but poorly on new data.

```python
#Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data: ● X = [1, 2, 3, 4, 5] ● Y = [2.1, 4.3, 6.1, 7.9, 10.2] Plot the regression line over the data points

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Prepare the data
# X must be a 2D array (column vector) for scikit-learn
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# 2. Initialize and fit the Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# 3. Predict values to draw the regression line
Y_pred = model.predict(X)

# 4. Plot the results
plt.scatter(X, Y, color='blue', label='Actual Data') # Data points
plt.plot(X, Y_pred, color='red', label='Regression Line') # Best-fit line
plt.title('Simple Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Optional: Print the calculated Slope and Intercept
print(f"Slope (Coefficient): {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")

```

```python
#Question 7: Fit a Multiple Linear Regression model on this sample data: ● Area = [1200, 1500, 1800, 2000] ● Rooms = [2, 3, 3, 4] ● Price = [250000, 300000, 320000, 370000] Check for multicollinearity using VIF and report the results


import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 1. Prepare the Data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Independent variables (X) and Dependent variable (y)
X = df[['Area', 'Rooms']]
y = df['Price']

# 2. Fit Multiple Linear Regression Model
model = LinearRegression()
model.fit(X, y)

print("--- Regression Results ---")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: Area={model.coef_[0]:.2f}, Rooms={model.coef_[1]:.2f}")
print(f"R^2 Score: {model.score(X, y):.4f}\n")

# 3. Check for Multicollinearity using VIF
# VIF requires a constant (intercept) to be added to the features
X_vif = X.copy()
X_vif['intercept'] = 1

vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i)
                   for i in range(len(X_vif.columns))]

print("--- VIF Results ---")
print(vif_data)


```

```python
#Question 8:Implement polynomial regression on the following data: ● X = [1, 2, 3, 4, 5] 3 ● Y = [2.2, 4.8, 7.5, 11.2, 14.7] Fit a 2nd-degree polynomial and plot the resulting curve

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# 1. Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# 2. Transform data for 2nd-degree polynomial (y = a*x^2 + b*x + c)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# 3. Fit Linear Regression model
model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

# 4. Plotting the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='2nd-Degree Polynomial Fit')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

# Display equation coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

```python
#Question 9: Create a residuals plot for a regression model trained on this data: ● X = [10, 20, 30, 40, 50] ● Y = [15, 35, 40, 50, 65] Assess heteroscedasticity by examining the spread of residuals.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

# Calculate Predicted Values and Residuals
y_pred = model.predict(X)
residuals = Y - y_pred

# Create Residual Plot
plt.figure(figsize=(8, 5))
plt.scatter(y_pred, residuals, color='blue', edgecolor='k')
plt.axhline(y=0, color='red', linestyle='--')  # Zero-error line
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values (Predicted Y)')
plt.ylabel('Residuals')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# Output Results for Reference
print(f"Fitted Values: {y_pred}")
print(f"Residuals: {residuals}")
```

Question 10:Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model

Ans:

As a data scientist tackling house price prediction, detecting heteroscedasticity and multicollinearity suggests that the ordinary least squares (OLS) regression assumptions are violated, leading to unreliable coefficient estimates and inefficient predictions. Here are the steps to address these issues and ensure a robust model:
1. Address Multicollinearity (High correlation between features)
Multicollinearity makes coefficients unstable and difficult to interpret (e.g., "area" and "number of rooms" are likely highly correlated).
Calculate Variance Inflation Factors (VIF): Identify features with high VIF (typically >10 or sometimes >5) to identify which variables are causing the issue.

Drop Redundant Features: If two variables are highly correlated (e.g., "Number of Bathrooms" and "Total Rooms"), drop one of them.
Feature Engineering (Combine Features): Instead of using multiple correlated features, create a combined feature, such as "Total Area" (Sum of living, basement, and garage) or "Room Density" (Area/Rooms).

Use Regularization (Ridge or Lasso Regression): Apply Ridge (l2) or Lasso (l1) regression. These techniques add a penalty term to the regression formula, which restricts coefficient size and mitigates the effect of multicollinearity.

Principal Component Analysis (PCA): Transform the features into a smaller set of uncorrelated components, though this reduces model interpretability.

2. Address Heteroscedasticity (Non-constant variance of residuals)
Heteroscedasticity, where error variance increases with price, results in invalid standard errors and p-values.
Log Transformation of the Dependent Variable: Apply a log transformation to the house price (



). This often shrinks the range of high-priced homes, reducing the fan-shape in residual plots.
Use Weighted Least Squares (WLS) Regression: If the error variance is related to an independent variable (e.g., area), assign smaller weights to observations with higher variance to reduce their impact.
Use Heteroscedasticity-Robust Standard Errors: Keep the OLS coefficients but use heteroscedasticity-consistent (HC) standard errors (e.g., White's estimator) to calculate valid confidence intervals and p-values.

3. Ensure a Robust Model and Validation
Outlier Detection and Treatment: Heteroscedasticity is often caused by extreme outliers. Use robust scaling or remove/cap extreme, non-representative properties.
Try Robust Regression Techniques: Use RANSAC or Theil-Sen estimators, which are less sensitive to outliers than OLS.
Cross-Validation: Use cross-validation to ensure the model generalizes well to new data rather than just fitting the noise in the training set.
Use Tree-Based Models: Consider Random Forest or Gradient Boosting (XGBoost/LightGBM). These models are robust to multicollinearity, handle non-linear relationships, and do not assume constant error variance.
By combining VIF reduction and regularization for multicollinearity, and using log-transformed targets or robust standard errors for heteroscedasticity, the resulting house price model will be more robust and accurate.
